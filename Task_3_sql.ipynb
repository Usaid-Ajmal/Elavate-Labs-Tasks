```python
%%sql

CREATE TABLE amzsales AS
SELECT "index" , "Order ID", "Date","Status","Sales Channel", "Category","Size","Courier Status","Qty","currency","Amount","ship-city","ship-state","ship-country"
FROM amazon_sales;
```

SELECT, WHERE, ORDER BY, GROUP BY


```python
%%sql
select "Category", count(*) as total
from amzsales
group by "Category"
having total >1000
order by total;
```

```python
%%sql
select "Order ID", "Category"
from amzsales
where "Amount">=1000
and "ship-city"=="MUMBAI"
limit 5;
```

#Create views for analysis

```python
%%sql
create view chennai_sales as
select *
from amzsales
where "ship-city" == "CHENNAI";


```

```python
%%sql
select * from chennai_sales where "Category"=="kurta" limit 5;
```

#Subqueries

```python
%%sql
select *
from amzsales
where "Amount">=(select AVG("Amount") from amzsales)
limit 5;
```
